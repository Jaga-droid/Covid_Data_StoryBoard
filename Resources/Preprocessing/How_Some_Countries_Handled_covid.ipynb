```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
whole_data=pd.read_csv("https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv")
```

```python
mydata=whole_data.copy()
```

```python
stringency=mydata[['date','location','stringency_index','tests_per_case','new_deaths','new_cases','hospital_beds_per_thousand','positive_rate','population_density']]
stringency['date']=pd.to_datetime(stringency['date'], format="%Y/%m/%d")
stringency
```

```python
stringency['year']=stringency['date'].dt.strftime('%Y')
```

```python
stringency=stringency[stringency['year'].isin(["2020","2021"])]
```

```python
stringency=mydata[['location','stringency_index','tests_per_case','new_deaths','new_cases','hospital_beds_per_thousand','positive_rate','population_density']].groupby(['location']).mean()
stringency=stringency.sort_values(by='stringency_index',ascending=False)
stringency['index']=[x for x in range(0,len(stringency))]
stringency['Country']=stringency.index
stringency=stringency.set_index('index')
stringency=stringency[stringency['Country'].isin(['Singapore','South Korea','India'])]
stringency=stringency.reset_index(drop=True)
stringency
```

```python
stringency.to_csv('../DataSet/Countriestacklingcovid.csv')
```
