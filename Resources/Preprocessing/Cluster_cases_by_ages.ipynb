```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
full_data=pd.read_csv("https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv")
```

```python
full_data.columns
```

```python
ages=full_data.copy()
```

```python
ages['>65_diff']=ages['aged_65_older']-ages['aged_70_older']
ages['>65']=ages['>65_diff']/100 *ages['total_cases']
ages['>70']=ages['aged_70_older'] /100 * ages['total_cases']

```

```python
ages['<65']=ages['total_cases']-ages['>65']-ages['>70']
```
